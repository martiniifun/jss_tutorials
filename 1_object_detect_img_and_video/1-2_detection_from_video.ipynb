내부적으로 많은 코드를 자동으로 처리해주지만

치명적인 문제점이 하나 있습니다.

동영상 처리가 완료될 때까지

동영상이 조금만 길어도 

메모리부족 현상이 발생하게 된다는 점입니다.

그래서 ultralytics에는
 
매 프레임의 추론결과를 generator에 리턴하는 옵션을 제공하고 있습니다.

> generator가 뭔지 모르신다면,
> `for i in range(100000000000000000000): print(i)`라는 코드와
> `for i in list(range(100000000000000000000)): print(i)`라는 코드를
> 한 번씩 실행해보시기만 해도 그 차이가 뭔지 어렴풋이 아시게 될 겁니다.
> 꼭 한 번 해보세요^^



stream=True 옵션으로 비디오를 추론하는

기본 뼈대는 아래와 같습니다.

> 앞으로 동영상을 다룰 때는 
> 항상 이 뼈대 위에 코드를 붙일 예정이므로 눈에 익혀 두시기 바랍니다.


```python
import cv2
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

video_path = "./video/관악IC.mp4"
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    success, frame = cap.read()

    if success:
        results = model(frame)
        #
        # 이 주석 부분에 results의 정보를 가지고 분석을 하거나
        # 로그를 남기는 등의 추가작업을 수행할 수 있습니다.
        # 지금은 추론결과를 프레임에 나타낼 뿐 
        # 아무 추가작업도 수행하지 않고 있습니다.
        #
        annotated_frame = results[0].plot()
        cv2.imshow("YOLOv8 Inference", annotated_frame)
        # ↑ YOLO가 아닌 opencv에서 창을 열었습니다.
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

위의 코드에서는 추론결과를 따로 동영상으로 저장하지는 않았습니다.
